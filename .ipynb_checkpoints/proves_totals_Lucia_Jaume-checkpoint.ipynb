```python
import numpy as np
import math
import pandas as pd
import os
from datetime import datetime
from typing import List, Dict, Tuple, Optional
import re
from pathlib import Path
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

import sys
sys.path.append('/home/jgarcia/Projects/mar_catala/src')
import HFradar_data as hf
import read_data as rd
```

```python
filepath = "/home/jgarcia/Projects/mar_catala/radial_reconstruction/data/real_data_radar/filtered_radials6_CREU_BEGU.txt"

df = pd.read_csv(filepath, sep=r'\s+', header=None,names=['LOND', 'LATD', 'VELU', 'VELV', 'BEAR', 'VELO', 'DIR', 'ANTENNA_FLAG'])
```

```python
antenna_flag_2 = df[df['ANTENNA_FLAG'] == 2]
antenna_flag_3 = df[df['ANTENNA_FLAG'] == 3]
```

```python
# BATHYMETRY
file_icatmar = '/home/jgarcia/Projects/mar_catala/bathy.nc'

lon_bat,lat_bat,h = rd.read_bath(file_icatmar)
```

```python
scale = 5
width = 0.003

fig, ax = plt.subplots(figsize=(10, 5), subplot_kw={'projection': ccrs.PlateCarree()})

#ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
ax.coastlines()
	
# Bathymetry
pcm = plt.pcolor(lon_bat,lat_bat,h, cmap = "Blues_r")

# Quiver UV
Q = ax.quiver(antenna_flag_2['LOND'],antenna_flag_2['LATD'],antenna_flag_2['VELU'],antenna_flag_2['VELV'],transform = ccrs.PlateCarree(),scale=scale,width=width,minlength = 0,color = 'green')
Q1 = ax.quiver(antenna_flag_3['LOND'],antenna_flag_3['LATD'],antenna_flag_3['VELU'],antenna_flag_3['VELV'],transform = ccrs.PlateCarree(),scale=scale,width=width,minlength = 0,color = 'red')
	
ax.quiverkey(Q,0.20, 0.92, 0.5,"0.5 m/s",labelpos='E')

x_min = 3.
x_max = 4.5
y_min = 41.5
y_max = 43


plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.tight_layout()
plt.show()
```

```python
file_tuv = "/home/jgarcia/Projects/mar_catala/radial_reconstruction/data/TOTL_CATS_2025_02_05_1000.tuv"

lon,lat,u,v,mod_vel,angle = hf.read_tuv(file_tuv);

# Convertir a arrays de float (equivalente a parse.(Float64, ...) en Julia)
lon_array = np.array([float(x) for x in lon])
lat_array = np.array([float(x) for x in lat])
u_array = np.array([float(x) for x in u]) / 100  # Dividir por 100
v_array = np.array([float(x) for x in v]) / 100  # Dividir por 100
mod_vel_array = np.array([float(x) for x in mod_vel]) / 100  # Dividir por 100
angle_array = np.array([float(x) for x in angle])

# Evitar outliers o velocidades no físicas (equivalente a mod_vel_array .<= 1.2)
valid_index = (
    (mod_vel_array <= 1.7) & 
    (lon_array <= 3.8) & 
    (lon_array >= 3.7) & 
    (lat_array <= 42.3) & 
    (lat_array >= 41.9)
)

# Filtrar todos los arrays usando el índice válido
lon_array = lon_array[valid_index]
lat_array = lat_array[valid_index]
u_array = u_array[valid_index]
v_array = v_array[valid_index]
mod_vel_array = mod_vel_array[valid_index]
angle_array = angle_array[valid_index]
```

```python
fig, ax = plt.subplots(figsize=(10, 5), subplot_kw={'projection': ccrs.PlateCarree()})

#ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
ax.coastlines()
	
# Bathymetry
pcm = plt.pcolor(lon_bat,lat_bat,h, cmap = "Blues_r")

# Quiver UV
Q = ax.quiver(lon_array,lat_array,u_array,v_array,transform = ccrs.PlateCarree(),scale=scale,width=width,minlength = 0,color = 'black')	
ax.quiverkey(Q,0.20, 0.92, 0.5,"0.5 m/s",labelpos='E')

x_min = 3.
x_max = 4.5
y_min = 41.5
y_max = 43


plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.tight_layout()
plt.show()
```
